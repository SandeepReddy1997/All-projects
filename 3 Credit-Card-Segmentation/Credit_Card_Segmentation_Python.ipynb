###### User Defined Function to catagorise the purchase behaviour in the data

```python
def purchaseType(data):
    if (data['ONEOFF_PURCHASES'] == 0) & (data['INSTALLMENTS_PURCHASES'] > 0) :
        return 'installment'
    if (data['ONEOFF_PURCHASES'] > 0) & (data['INSTALLMENTS_PURCHASES'] == 0) :
        return 'one_off'
    if (data['ONEOFF_PURCHASES'] > 0) & (data['INSTALLMENTS_PURCHASES'] > 0) :
        return 'both'
    if (data['ONEOFF_PURCHASES'] == 0) & (data['INSTALLMENTS_PURCHASES'] == 0) :
        return 'not any'
```

```python
data_KPI['Purchase_Type'] = data.apply(purchaseType, axis = 1)
```

# Cluster Analysis

```python
# Find the Optimal number of clusters using elbow method

from sklearn.cluster import KMeans

# Estimate Optimum number of clusters
cluster_range = range(1, 20)
cluster_errors = []

for noOfClusters in cluster_range:
    clusters = KMeans(noOfClusters).fit(data)
    cluster_errors.append(clusters.inertia_)
    
cluster_DF = pd.DataFrame( {"noOfClusters": cluster_range, "cluster_errors": cluster_errors})
cluster_DF
```

### Plotting Elbow Graph

```python
plt.figure(figsize=(18, 6))
plt.plot(cluster_DF.noOfClusters, cluster_DF.cluster_errors, marker = 'o')
plt.xlabel('K')
plt.ylabel('ERROR')
plt.title('Elbow Graph')
```

#### Creating the KMeans Model

```python
# Cluster Size
cluster_size = 4
```

```python
# Creating the Model
kmeans_model = KMeans(n_clusters = cluster_size).fit(data)
```
